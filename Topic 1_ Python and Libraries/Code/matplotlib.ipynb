```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
housing = pd.read_csv('housing.csv') 
```

# Histogram important give us distibution 


```python
price = housing.median_house_value
price
price.hist()
```

```python
plt.hist(price)
plt.grid()
plt.show()
```

```python
plt.hist(np.log(price)) # the previuse one graph look no good we need to use 'log'
plt.grid()
plt.show()
```

## 3. line plot


```python
x= np.arange(10)
y = 2 * x + np.random.rand(10)* 2
y2 = 2 * x + np.random.rand(10)*3
```

```python
plt.plot(x,y, linestyle = 'dashed', marker = 'o', color='red')
plt.plot(x,y2, linestyle = 'dotted', marker = '^', color='blue')
plt.grid()
plt.show()
```
