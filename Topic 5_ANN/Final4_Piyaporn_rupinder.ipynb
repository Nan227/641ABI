```python
import tensorflow as tf

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.shape, test.shape
```

```python
# Add logSalePrice column
train['logSalePrice'] = np.log(train['SalePrice'])
test['logSalePrice'] = np.log(test['SalePrice'])

# Drop the first column from both datasets
train = train.iloc[:, 1:]
test = test.iloc[:, 1:]
```

```python
# Assuming your DataFrame is named df
train['StreetPave'] = train['Street'].apply(lambda x: 1 if x == 'Pave' else 0)
test['StreetPave'] = test['Street'].apply(lambda x: 1 if x == 'Pave' else 0)

# Drop the Street column
train = train.drop(columns=['Street'])
test = test.drop(columns=['Street'])
```

```python
# Fix the CentralAir column
train['CentralAirY'] = train['CentralAir'].apply(lambda x: 1 if x == 'Y' else 0)
test['CentralAirY'] = test['CentralAir'].apply(lambda x: 1 if x == 'Y' else 0)

# Drop the CentralAir column
train = train.drop(columns=['CentralAir'])
test = test.drop(columns=['CentralAir'])
```

```python
# Fix LotShape column in both datasets (add 3 new columns) instead of 4 )
# IR1, IR2, IR3,  (NO! Reg)
train['LotShapeIR1'] = train['LotShape'].apply(lambda x: 1 if x == 'Y' else 0) 
test['LotShapeIR1'] = test['LotShape'].apply(lambda x: 1 if x == 'Y' else 0)

train['LotShapeIR2'] = train['LotShape'].apply(lambda x: 1 if x == 'Y' else 0)
test['LotShapeIR2'] = test['LotShape'].apply(lambda x: 1 if x == 'Y' else 0)

train['LotShapeIR3'] = train['LotShape'].apply(lambda x: 1 if x == 'Y' else 0)
test['LotShapeIR3'] = test['LotShape'].apply(lambda x: 1 if x == 'Y' else 0)

# Drop the LotShape column
train = train.drop(columns=['LotShape'])
test = test.drop(columns=['LotShape'])
```
